```python
#creating a dataframe using pandas
ds=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
ds['target']=iris.target
```

```python
ds1=ds[:50]  #first 50 features/rows belongs to target 0
ds2=ds[50:100] # 50-99 rows/features belongs to target 1
ds3=ds[100:] # 100-149 rows/features belongs to target 2
```

```python
ds.columns
```

```python
import matplotlib.pyplot as plt
plt.xlabel('sepal length')
plt.ylabel('sepal width')

plt.scatter(ds1['sepal length (cm)'],ds1['sepal width (cm)'],color='green')
plt.scatter(ds2['sepal length (cm)'],ds2['sepal width (cm)'],color='red')
plt.scatter(ds3['sepal length (cm)'],ds3['sepal width (cm)'],color='blue')
```

```python
ds.columns
```

```python
ds.shape
```

```python
x=ds[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
y=ds[['target']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

```python
y_predicted=model.predict(x_test)
y_predicted
```

```python
y_test
```

```python
model.score(x_test,y_test)
```

```python
# confusion_matrix=performance of the model
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
plt.xlabel('actual')
plt.ylabel('predicted')
```

```python
x_test.shape
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predicted))  
```

```python
# # confusion_matrix= performance of the model
# # classification_report= performace evaluation metric
# # precision= positive prediction value= out of total predicted positive result how many results are actually positive
# # recall= True positive Rate/sensitivity- how many positive values out of all predictions
# # f1 score= mean/avg of precision and recall
# # support=no. of samples supported for each class.


# TP=True Positive +ve
# TN=True Negative -ve
# FP=False Positive -ve
# FN=False Negative +ve


# 1000 samples 500 Yes and 500 No ==> balanced dataset
#              600 Yes and 400 No ==> balanced dataset
#              900 Yes and 100 No ==> imbalanced dataset
        
# balanced dataset===> accuracy

# imbalanced dataset==> precision recall and f1 score

# when FP rates must be reduced ==> precision ex: spam detection
# when FN rates must be reduced ==> recall ex: cancer or not
```
